```python
import pandas as pd
import random
```

```python
data = pd.read_csv('source.txt', sep='\t')
print(data)
```

```python
names = [x.lower() for x in data.name.tolist()]
print(names)
```

```python
mergable = {}
for name in names:
    a = name[0]
    b = name[-1]
    key = a+b
    mergable[key] = mergable.get(key, []) + [name]

import json
print(json.dumps(mergable, indent=2))
```

```python
starts = {}
ends = {}
for key in mergable.keys():
    s, e = key[0], key[-1]
    starts[s] = starts.get(s, []) + [key]
    ends[e] = ends.get(e, []) + [key]

print(json.dumps(starts, indent=2))
print(json.dumps(ends, indent=2))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
```

```python
# add nodes
for key, countries in mergable.items():
    G.add_node(key, countries=countries, count=len(countries))

# add edges
for key in mergable.keys():
    end = key[-1]
    children = starts.get(end, [])
    
    for child in children:
        G.add_edge(key, child)
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()
```

```python
# Find sources / sinks
sources = []
sinks = []

in_edge_counts = []
out_edge_counts = []

for node in G.nodes():
    data = G.nodes[node]
    outs = G.out_edges(node)
    ins = G.in_edges(node)
    
    in_edge_counts.append(len(ins))
    out_edge_counts.append(len(outs))
#     print(node, len(ins), len(outs))
    
    if len(ins) == 0:
        sources.append(node)

    if len(outs) == 0:
        print(f'Found sink node: {node}')
        sinks.append(node)

print(len(sources), min(out_edge_counts), max(out_edge_counts))
print(len(sinks), min(in_edge_counts), max(in_edge_counts))
```

```python
print(f'Sources: {sources}')
print(f'Sinks: {sinks}')
print(f'Total num of nodes: {len(G.nodes)}')
```

```python
# Make some pointers and get some counts
N = 100
M = 20
K = 5
counts = {}

for k in range(K):
    # pointers = [random.choice(list(G.nodes)) for _ in range(N)]
    pointers = list(G.nodes)
    for i in range(M):
        new_pointers = []
        for p in pointers:
            counts[p] = counts.get(p, 0) + 1
            succ = list(G.successors(p))
            p2 = random.choice(succ) if succ else p
            new_pointers.append(p2)
        pointers = new_pointers

print(f'Finished with these counts: ')
print(json.dumps(counts, indent=2))
```

```python
topo_sorted = sorted(G.nodes, key=lambda x: counts.get(x, 0))
topos = [(x, counts.get(x, 0)) for x in topo_sorted]
print(topos)
```

```python
topo_positions = {node: (i, -val) for i, (node, val) in enumerate(topos)}
nx.draw(G, pos=topo_positions, with_labels=True)
plt.show()
```

```python
from pyvis.network import Network
net = Network(notebook=True, directed=True)

for node in G.nodes:
    net.add_node(node)
    
for node in G.nodes:
    for child in G.successors(node):
        net.add_edge(node, child)
        
net.show('example.html')
```
